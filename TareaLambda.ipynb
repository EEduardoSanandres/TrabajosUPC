Implementar 4 métodos:

*   *Insertar al inicio*
*   *Buscar en posición*
*   *Eliminar al final*
*   *Ordenar ascendente*












```python
%%writefile Biblioteca.hpp
#include <iostream>
#include <functional>
#include <cstdlib>// srand, rand
#include <ctime>//time
using std::function;
using std::cout;
using std::cin;
//lambdas
void f1(function<void(int*,int)> funcion,int* a, int b){return funcion(a,b);}
int* f2(function<int*(int*,int)> funcion,int* a, int b){return funcion(a,b);}
//Funciones
void listar(int* a, int n) {
    for(int i = 0; i<n; ++i){
        cout << i << " -> " << a[i] << "\n";
    }
}
void CrearArreglo(int* a, int n) {
  for(int i=0;i<n;++i){
    a[i]=rand()%101+100;//rango de 100 a 200
  }
  listar(a,n);
}
int* InsertarInicio(int* a, int n) {
  int* aux = new int[n + 1];
  int v;
  cin >> v;
  aux[0] = v;
  for(int i=0;i<n;++i){
      aux[i+1]=a[i];    
  }
  a = new int[n + 1];
  for(int i=0;i<n+1;++i){
      a[i]=aux[i];    
  }
  return a;
}
void ordenar(int* a, int n){
  for(int i=0;i<n-1;++i){
    for(int j=i+1;j<n;++j){
      if(a[i]<a[j]){
        int aux=a[i];
        a[i]=a[j];
        a[j]=aux;
      }
    }
  }
  listar(a,n);
}
void buscar(int* a, int n) {
    int aux;
    cout << "Posicion que desee buscar: "; cin >> aux;
    for(int i=0;i<n;++i){
        if(i == aux) {
            cout << "El elemento en la posicion " << aux << " es: " << a[i] << "\n";
        }
    }
}
int* eliminar(int* a, int n) {
  int* aux = new int[n-1];
  for(int i=0;i<n-1;++i){
      aux[i]=a[i];    
  }
  a = new int[n - 1];
  for(int i=0;i<n-1;++i){
      a[i]=aux[i];    
  }
  return a;
}
```

```python
%%writefile lambda.cpp
#include "Biblioteca.hpp"
using std::cout;
using std::cin;
int main() {
    int n = 1;
    srand(time(NULL));
    cout << "Ingresar el numero de elemntos: \n"; cin >> n;
    int* arreglo = new int[n];
    f1(CrearArreglo,arreglo, n);
    cout<< "INSERTAR AL INICIO \n";
    arreglo = f2(InsertarInicio,arreglo, n);
    n++;
    f1(listar,arreglo, n);
    cout<< "ORDENAR \n";
    f1(ordenar,arreglo, n);
    cout<< "BUSCAR \n";
    f1(buscar,arreglo, n);
    cout<< "ELIMINAR \n";
    arreglo = f2(eliminar,arreglo, n);
    n--;
    f1(listar,arreglo, n);
    return 0;
}
```
