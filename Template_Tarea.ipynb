```python
%%writefile Biblioteca.hpp
//Definicion de bibliotecas, clases y funciones
#include <iostream>
#include <cstdlib>// srand, rand
#include <ctime>//time

using std::cout;
using std::cin;

template <typename Generico>
class MiArreglo{
    Generico* a;
    size_t id;
    size_t n;
    public:
    MiArreglo() {
        n = 10;
        a = new Generico[n];
        id = 0;
    }
void insertar(Generico g) {
    if(id<n)
      a[id++]=g;
}
void listar() {
    for(size_t i = 0; i<n; ++i){
        cout << i << " -> " << a[i] << "\n";
    }
}
void mayor(){
  Generico m=a[0];
  short p=-1;
  for(size_t i=1;i<n;++i){
    if(a[i]>m){
      m=a[i];
      p=i;
    }
  }
  cout<<"El mayor valor es "<<m<<" y se encuentra en la posicion "<<p<<"\n";
}
void menor(){
  Generico m=a[0];
  short p=0;
  for(size_t i=1;i<n;++i){
    if(a[i]<=m){
      m=a[i];
      p=i;
    }
  }
  cout<<"El menor valor es "<<m<<" y se encuentra en la posicion "<<p<<"\n";
}
void ordenar(){
  for(int i=0;i<n-1;++i){
    for(int j=i+1;j<n;++j){
      if(a[i]>a[j]){
        Generico aux=a[i];
        a[i]=a[j];
        a[j]=aux;
      }
    }
  }
}
};
```

```python
%%writefile ejemplo.cpp
#include "Biblioteca.hpp"
#include <iostream>
#include <cstdlib>// srand, rand
#include <ctime>//time
int main() {
    MiArreglo<short>* arr1 = new MiArreglo<short>();
    MiArreglo<float>* arr2 = new MiArreglo<float>();
    MiArreglo<long>* arr3 = new MiArreglo<long>();
    for(size_t i = 0;i<10;++i){
        arr1->insertar(rand()%101+100);
        arr2->insertar((rand()%101+100)*1.654);
        arr3->insertar((rand()%101+100)*1000000);
    }
    arr1->listar(); cout << "\n";
    arr2->listar(); cout << "\n";
    arr3->listar(); cout << "\n";
    arr1->mayor(); cout;
    arr2->mayor(); cout;
    arr3->mayor(); cout;
    arr1->menor(); cout;
    arr2->menor(); cout;
    arr3->menor(); cout;
    arr1->ordenar(); cout;
    arr2->ordenar(); cout;
    arr3->ordenar(); cout;
    arr1->listar(); cout << "\n";
    arr2->listar(); cout << "\n";
    arr3->listar(); cout << "\n";
}
```
